```python
import os
import pandas as pd

path ="./Data"

filename_read = os.path.join(path,"auto-mpg.csv")
df = pd.read_csv(filename_read,na_values=['NA', '?'])

print(df)
```

```python
# Strip non-numerics 

df = df.select_dtypes(include=['int','float'])
print (df)
```

```python
headers = list(df.columns.values)
print(headers)
```

```python
# find the mean, variance, and standard deveation
fields = []
for field in headers:
    fields.append( {
        'name' : field,
        'mean' : df[field].mean(),
        'var'  : df[field].var(), 
        'sdev' : df[field].std()
    })
for field in fields:
    print(field)
```
